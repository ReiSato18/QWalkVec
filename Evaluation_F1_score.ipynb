```python
def F1_score(df,train_size):
    micro = []
    macro = []
    #karate, webkbの場合node_label[0]
    #node_labelのカラムはデータによって違うので注意
    X_train = df
    #y_train = pd.DataFrame(list(node_label[0]))
    y_train = df_label["y"]
    
    model = LogisticRegression(random_state=0, fit_intercept=False, multi_class='auto', solver='liblinear')
    for i in range(10):
        train_x, valid_x, train_y, valid_y = train_test_split(X_train, y_train, test_size=1-train_size)
        #model = LogisticRegression(random_state=0, fit_intercept=True, multi_class='auto', solver='liblinear')
        #model = LogisticRegression(random_state=0, fit_intercept=True, multi_class='ovr', solver='liblinear')
        model.fit(train_x, train_y.values.ravel())
        predicted = model.predict(valid_x)

        micro.append(f1_score(y_true=valid_y, y_pred=predicted, average="micro"))
        macro.append(f1_score(y_true=valid_y, y_pred=predicted, average="macro"))
        #precision.append(precision_score(y_true=valid_y, y_pred=predicted))
        #recall.append(recall_socre(y_true=valid_y, y_pred=predicted))
    return micro,macro

def per_step_F1_score_QW(train_size,df):
    micro_f1 = []
    err_micro_f1 = []
    macro_f1 = []
    err_macro_f1 = []
    #itrと同値
    num_column = len(df.iloc[0,:])
    if num_column == 1:
        result = F1_score(df ,train_size)

        micro_f1.append(np.mean(result[0]))
        macro_f1.append(np.mean(result[1]))
        #err_micro_f1.append(statistics.pstdev(result[0]))
        #err_macro_f1.append(statistics.pstdev(result[1]))
    else:
        for d in range(1,num_column):

            result = F1_score(df.iloc[:,0:d],train_size)

            micro_f1.append(np.mean(result[0]))
            macro_f1.append(np.mean(result[1]))
            err_micro_f1.append(statistics.pstdev(result[0]))
            err_macro_f1.append(statistics.pstdev(result[1]))
    
    return micro_f1,err_micro_f1,macro_f1,err_macro_f1
    #return micro_f1,macro_f1
```

# Grid search

```python
def F1_score(df,train_size):
    micro = []
    macro = []
    #karate, webkbの場合node_label[0]
    #node_labelのカラムはデータによって違うので注意
    X_train = df
    #y_train = pd.DataFrame(list(node_label[0]))
    y_train = df_label["y"]
    
    model = LogisticRegression(random_state=0, fit_intercept=False, multi_class='auto', solver='liblinear')
    for i in range(20):
        train_x, valid_x, train_y, valid_y = train_test_split(X_train, y_train, test_size=1-train_size)
        
        model.fit(train_x, train_y.values.ravel())
        predicted = model.predict(valid_x)

        micro.append(f1_score(y_true=valid_y, y_pred=predicted, average="micro"))
        macro.append(f1_score(y_true=valid_y, y_pred=predicted, average="macro"))
    
    return micro,macro

def per_step_F1_score_QW(train_size,df):
    micro_f1 = []
    err_micro_f1 = []
    macro_f1 = []
    err_macro_f1 = []
    #itrと同値
    num_column = len(df.iloc[0,:])
    if num_column == 1:
        result = F1_score(df ,train_size)

        micro_f1.append(np.mean(result[0]))
        macro_f1.append(np.mean(result[1]))
    else:
        for d in range(1,num_column):

            result = F1_score(df.iloc[:,0:d],train_size)

            micro_f1.append(np.mean(result[0]))
            macro_f1.append(np.mean(result[1]))
    
    return micro_f1,macro_f1
```

```python
tR = 0.5
graph = label_set[0]
def make_result(itr):
    micro_max = dict()
    macro_max = dict()
    for p in [0.25,0.50,1.0,2.0,4.0]:
        for q in [0.25,0.50,1.0,2.0,4.0]:
            df = dataset_table(p,q,graph,"superposition",itr)
            result = per_step_F1_score_QW(tR,df)
            micro_max[p,q] = np.max(result[0])
            macro_max[p,q] = np.max(result[1])
        print("---->")
    
    return micro_max, macro_max
```
